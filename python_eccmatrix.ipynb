```python
# eccentricity matrix
import numpy as np
from numpy.linalg import eig
import networkx as nx
import matplotlib.pyplot as plot
def eccmt(G,n):
    eccm = np.zeros((n,n))
    for i in range(0,n-1):
        for j in range(i+1,n):
            if nx.shortest_path_length(G,i,j) == min(nx.eccentricity(G,i),nx.eccentricity(G,j)):
                eccm[i][j]=nx.shortest_path_length(G,i,j)
    for i in range(1,n):
        for j in range(0,i):
            eccm[i][j] = eccm[j][i]
    return eccm
T = nx.Graph()
T.add_nodes_from([0,1,2,3])
T.add_edges_from([(1, 2), (1, 3),(0, 3),(2, 3)])
n = T.number_of_nodes()
print(eccmt(T,n),max(eig(eccmt(T,n))[0]))
nx.draw(T, with_labels=True)
```
