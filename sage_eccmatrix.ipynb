```python
def eccms(G,n):
    eccm=matrix(QQ, n)
    for i in range(0,n-1):
        for j in range(i+1,n):
            if G.distance(i,j)==min(G.eccentricity(i),G.eccentricity(j)):
                eccm[i,j]=G.distance(i,j)
    for i in range(1,n):
        for j in range(0,i):
            eccm[i,j]=eccm[j,i]
    return eccm
m=5
g = graphs.CompleteGraph(m)
p = graphs.PathGraph(m)
print(eccms(g,m),'\n',eccms(p,m))
g.show()
```
